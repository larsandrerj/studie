```python
import numpy as np
import matplotlib.pyplot as plt

# Sett hvilke x-verdier du vil plotte for
x = np.linspace(0,5,100)

# Sett verdien for parameteren alpha
alpha = 1

# Beregn kumulativ fordelingsfunksjon
def F_X(x, alpha):
    return 1- np.exp(-x**2/alpha)

plt.plot(x, F_X(x, alpha), color="red")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.show()
```

```python
#Oppgave 7a)

def f_X(x, alpha):
    return ((2*x)/alpha) * np.exp((-x**2)/alpha)


x = np.linspace(0, 5)


plt.plot(x, f_X(x, alpha), color="red")
plt.xlabel("x")
plt.ylabel("f_X(x)")
plt.show()
```

```python
#Oppgave 7b)

def f_Z(z, alpha):
    return (4/alpha)*z*np.exp(-(z**2)/alpha) - (4/alpha)*z*np.exp(-(2*z**2)/alpha)

plt.plot(x, f_X(x, alpha), color="red")
plt.plot(x, f_Z(x, alpha), color="green")
plt.xlabel("x")
plt.ylabel("")
plt.show()
```
