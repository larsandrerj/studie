```python
import numpy as np 
import matplotlib.pyplot as plt 

n = 100000
#n = 10

x = np.random.normal(0, 1, n)

mu = 1
sigma = 2
a = 2
b = 0.5

y = a*x + b

z = np.random.normal(mu, sigma, n)


plt.hist(y, bins=100, density=True, alpha=0.6, color='red')
plt.hist(z, bins=100, density=True, alpha=0.6, color='green')
plt.show()
```
